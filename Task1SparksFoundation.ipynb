```python

import pandas as pd

# Read the CSV file into a DataFrame
data = pd.read_csv('/content/sample_data/Sparks.csv')

# Converting the data array into a list of lists
data_list = data.values.tolist()

# Spliting each list on the comma separator
for i in range(len(data_list)):
    data_list[i] = data_list[i][0].split(',')
```

```python
data=pd.DataFrame(data_list)
```

```python
data.head(10)
```

```python

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# Sample data
x = data[0]
x1=x.to_numpy().reshape(-1,1)
y = data[1]
y1=y.to_numpy().reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.2, random_state=42)
X= X_train.reshape(-1, 1)
# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)
value = np.array(9.25).reshape(-1, 1)
# Predict using the model
ypred = model.predict(value)

print('Predicted value for 9.25 hours of study',ypred)
fig, axis = plt.subplots(figsize=(12, 15))
plt.plot(x, y,color='red',marker="*",linestyle='--')

# Add labels and title
plt.xlabel('Hours of Study')
plt.ylabel('Scores')
plt.title('Linear Regression Chart')

# Display the chart
plt.show()
```
